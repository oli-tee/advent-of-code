```python
def solve(data, delta_row, delta_col):
    row = 0
    col = 0
    nrow = len(data)
    ncol = len(data[0])
    tree_count = 0
    while row + delta_row < nrow:
        row = row + delta_row
        col = (col + delta_col) % ncol
        if data[row][col] == '#':
            tree_count += 1

    return tree_count


test_input_1 = [
    '....', '....', '....'
]

assert solve(test_input_1, 1, 3) == 0

test_input_2 = [
    '....', '...#', '....'
    
]
assert solve(test_input_2, 1, 3) == 1


print('all tests passed')
```

```python
solve(data, 1, 3)
```

```python
solve(data, 1, 1) * solve(data, 1, 3) * solve(data, 1, 5) * solve(data, 1, 7) * solve(data, 2, 1)
```
